```python
import os, glob
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("Finaldataset3.0.csv")
dataset.head()
```

```python
sns.pairplot(data = dataset, height = 2);
```

```python
sucide_rate_mean = dataset[["country", "Suicide_rate"]].groupby(by = "country").mean()
sucide_rate_mean
```

```python
sucide_rate_mean = sucide_rate_mean.sort_values(by = "Suicide_rate", ascending = False)
sucide_rate_mean["country"] = sucide_rate_mean.index
sucide_rate_mean.reset_index(drop = True, inplace = True)
sucide_rate_mean
```

```python
plt.figure(figsize = (15, 7.5))
sns.barplot(data = sucide_rate_mean, x = "country", y = "Suicide_rate")
plt.xticks(rotation  = 90);
```
